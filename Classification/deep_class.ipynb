```python
X = df_cancer.drop('benign_0__mal_1' , axis =1).values
y =df_cancer['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,
                                                      test_size=0.25, 
                                                      random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30 , activation= 'relu'))
model.add(Dense(15 , activation= 'relu'))

#BINARY CLASSIFICATION so we will choose sigmoid
model.add(Dense(1 , activation= 'sigmoid'))

model.compile(loss='binary_crossentropy' , optimizer = 'adam')
```

```python
model.fit(x =X_train , y= y_train , epochs =600,
          validation_data= (X_test , y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.head()
```

```python
losses.plot()
```

### it is clear that  the model is overfitting
So ae are goin to recreate the model

```python
model = Sequential()
model.add(Dense(30 , activation= 'relu'))
model.add(Dense(15 , activation= 'relu'))

#BINARY CLASSIFICATION so we will choose sigmoid
model.add(Dense(1 , activation= 'sigmoid'))

model.compile(loss='binary_crossentropy' , optimizer = 'adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor= 'val_loss' , mode ='min' , verbose= 1 , patience =25)
```

```python
model.fit(x =X_train , y= y_train , epochs =600,
          validation_data= (X_test , y_test),
          callbacks= [early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(30 , activation= 'relu'))
## half the neurons in each batches in this layers are going to turned off
model.add(Dropout(0.5))
model.add(Dense(15 , activation= 'relu'))

#BINARY CLASSIFICATION so we will choose sigmoid
model.add(Dense(1 , activation= 'sigmoid'))

model.compile(loss='binary_crossentropy' , optimizer = 'adam')
```

```python
model.fit(x =X_train , y= y_train , epochs =600,
          validation_data= (X_test , y_test),
          callbacks= [early_stop])
```

```python
new_model_loss = pd.DataFrame(model.history.history)
```

```python
new_model_loss.plot()
```

```python
predictions = model.predict(X_test)
```

```python
predicted_classes = (predictions > 0.5).astype("int32")
```
