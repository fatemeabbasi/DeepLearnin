```python
house_df['date'] = pd.to_datetime(house_df['date'])
```

```python
house_df['date']
```

```python
house_df['year']= house_df['date'].apply(lambda date: date.year)
house_df['month']= house_df['date'].apply(lambda date: date.month)
house_df['day']= house_df['date'].apply(lambda date: date.day)
```

```python
house_df.head()
```

```python
plt.figure(figsize =(10,6))
sns.boxplot(x ='month' , y = 'price' , data= house_df)
```

```python
house_df.groupby('month').mean()['price'].plot()
```

```python
house_df.groupby('year').mean()['price'].plot()
```

```python
house_df.groupby('day').mean()['price'].plot()
```
